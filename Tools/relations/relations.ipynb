# ROI APPROXIMATION
The dependency on the energy looks like it is linear. Thus, a line is fitted for dataset and coefficients are used in approximation. 

### ROIL - lower boundary

```python
param="roih"
name = param+"mm_all"
plt.xlabel("Energy")
plt.ylabel("Pixel value")
plt.title(name)
for wd in wds:
    x, y = list(zip(*mm_s[param][wd]))
    # plt.plot(x,y, label=wd+"mm")
    # plt.title(wd+"mm")

    coef_a, coef_b = np.polyfit(x,y,1)

    print(wd, "mm", "a: ", coef_a, ", b: ", coef_b)
    x= np.array(x,np.float32)
    plt.plot(x, coef_a*x+coef_b, label=wd+"mm")

    # fig = plt.figure()

# plt.close(fig)

plt.legend()
plt.show()
```

### ROIH approximation

```python
param = "roih"
name = param+"mm_all"

plt.title(name)
for wd in wds:
    x, y = list(zip(*mm_s[param][wd]))
    plt.plot(x, y, label=wd+"mm")
    plt.title(wd+"mm")

    coef_a, coef_b = np.polyfit(x, y, 1)

    print(wd, "mm", "a: ", coef_a, ", b: ", coef_b)
    x = np.array(x, np.float32)
    plt.plot(x, coef_a*x+coef_b)

    fig = plt.figure()

plt.close(fig)

plt.legend()
plt.show()
```
